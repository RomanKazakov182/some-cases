### Задача 1. Поработаем с простыми визуализациями

```python
##############################################   Задание 1.1   #################################################################

######## Постройте график с помощью plt.scatter, где по горизонтальной оси будут значения Х, а по вертикальной оси - значения Y.
######## Добавьте название графика "Мой первый график"
######## Добавьте подпись горизонтальной оси "Ряд Х"
######## Добавьте подпись вертикальной оси "Ряд Y"
######## Добавьте сетку на график с помощью функции grid

from matplotlib import pyplot as plt


x = [1,2,3,4,5,6,7,8,9]
y = [8,5,6,10,18,26,18,22,23]

plt.scatter(x, y)
plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')
plt.grid()
plt.show()
```

```python
##############################################   Задание 1.2   #################################################################

######## Постройте график с помощью plt.plot, где по горизонтальной оси будут значения Х, а по вертикальной оси - значения Y.
######## Повторите все те же самые действия по оформлению графика, как в прошлом пункте.
######## Установите размер графика 7*7 с помощью параметра figsize функции figure

fig = plt.figure(figsize=(7, 7))
plt.plot(x, y)
plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')
plt.grid()
plt.show()
```

```python
##############################################   Задание 1.3   #################################################################

##### Добавьте на визуализацию из прошлого пункта сплошную вертикальную линию красного цвета в точке 6 
##### Добавьте также пунктирную вертикальную линию красного цвета в точке 2
##### (чтобы сделать линию пунктирной, воспользуйтесь аргументом linestyle)
##### Сделайте кривую y(x) зеленого цвета
##### Добавьте легенду для максимума и минимума в правый нижний угол
##### Для этого загуглите как работает функция axvline

import numpy as np

max_idx = np.argmax(y)
min_idx = np.argmin(y)
x_max = x[max_idx]
x_min = x[min_idx]

fig = plt.figure(figsize=(7, 7))
plt.plot(x, y, color = 'g')
plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')
plt.grid()
plt.axvline(x = x_max, color = 'r', label = f'Максимум y (х = {x_max: .2f})')
plt.axvline(x = x_min, color = 'r', linestyle='--', label = f'Минимум y (х = {x_min: .2f})')
plt.legend(loc = 'lower right')
plt.show()
```

```python
##############################################   Задание 1.4   #################################################################

##### Добавьте на график еще одную кривую z = [6,9,13,13,16,12,20,18,17] и сделайте ее фиолетовго цвета
##### Добавьте сплошную вертикальную линию черного цвета в точке 7
##### Добавьте пунктирную вертикальную линию черного цвета в точке 1
##### Подпишите их на легенде как Максимум Z и Минимум Z соответственно

z = [6,9,13,13,16,12,20,18,17]

max_idx_y = np.argmax(y)
min_idx_y = np.argmin(y)
x_max_y = x[max_idx_y]
x_min_y = x[min_idx_y]

max_idx_z = np.argmax(z)
min_idx_z = np.argmin(z)
x_max_z = x[max_idx_z]
x_min_z = x[min_idx_z]

fig = plt.figure(figsize=(7, 7))
plt.plot(x, y, color = 'g')
plt.plot(x, z, color = 'violet')
plt.title('Мой первый график')
plt.xlabel('Ряд Х')
plt.ylabel('Ряд Y')
plt.grid()
plt.axvline(x = x_max_y, color = 'r', label = f'Максимум y (х = {x_max_y: .2f})')
plt.axvline(x = x_min_y, color = 'r', linestyle='--', label = f'Минимум y (х = {x_min_y: .2f})')
plt.axvline(x = x_max_z, color = 'black', label = f'Максимум z (х = {x_max_z: .2f})')
plt.axvline(x = x_min_z, color = 'black', linestyle='--', label = f'Минимум z (х = {x_min_z: .2f})')
plt.legend(loc = 'lower right')
plt.show()
```

```python
##############################################   Задание 1.6   #################################################################

##### Добавьте на правую ось гистограмму, которая будет показывать разницу между Y и Z
##### Сделайте эту гистограмму полупрозрачной с помощью аргумента alpha
##### Добавьте подпись для гистограммы на отдельный элемент легенды.
##### Расположите основную легенду справа-снизу, а легенду вспомогательной оси справа-сверху

########  Hint: используйте twinx() для создания дополнительной оси

max_idx_y = np.argmax(y)
min_idx_y = np.argmin(y)
x_max_y = x[max_idx_y]
x_min_y = x[min_idx_y]

max_idx_z = np.argmax(z)
min_idx_z = np.argmin(z)
x_max_z = x[max_idx_z]
x_min_z = x[min_idx_z]

fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(x, y, color = 'g')
ax.plot(x, z, color = 'violet')
ax1 = ax.twinx()
c = [y - z for y, z in zip(y, z)]
ax1.bar(x, c, alpha = .25, color = 'grey', label = 'Разница')
ax.set_title('Мой первый график')
ax.set_xlabel('Ряд Х')
ax.set_ylabel('Ряд Y')
ax.grid()
ax.axvline(x = x_max_y, color = 'r', label = f'Максимум y (х = {x_max_y: .2f})')
ax.axvline(x = x_min_y, color = 'r', linestyle='--', label = f'Минимум y (х = {x_min_y: .2f})')
ax.axvline(x = x_max_z, color = 'black', label = f'Максимум z (х = {x_max_z: .2f})')
ax.axvline(x = x_min_z, color = 'black', linestyle='--', label = f'Минимум z (х = {x_min_z: .2f})')
ax.legend(loc = 'lower right')
ax.legend(loc = 'lower right')
ax1.legend(loc = 'upper right')
plt.show()
```

### Задание 2. Поработаем с визуализацией датасета

```python
##############################################   Задание 2.1   #################################################################

##### Импортните csv-файл под названием Purchase_List
##### Приведите date_purchase к формату даты с помощью функции to_datetime
##### Создайте колонку датафрейма с месяцесем оплат (числом, от 1 до 12)
##### Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
##### Постройте визуализацию: для каждого месяца покупки по левой оси постройте сумму оплат, а по правой - количество оплат
##### Не забудьте оформление графика (заглавие, подпись осей, раздельная легенда, сетка)

import pandas as pd
import datetime as dt

purchase_list = pd.read_csv('Purchase_List.csv')
purchase_list['date_purchase'] = pd.to_datetime(purchase_list['date_purchase'])
purchase_list['month_pay'] = purchase_list['date_purchase'].dt.month
df_group_month = purchase_list.groupby('month_pay').agg(cnt_pay = ('amt_pay', 'count'), sum_pay = ('amt_pay', 'sum')).reset_index()
df_group_month

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(df_group_month['month_pay'], df_group_month['sum_pay'], color = 'g', alpha = .9, label = 'Выручка в мес')
ax1 = ax.twinx()
ax1.bar(df_group_month['month_pay'], df_group_month['cnt_pay'], color = 'gold', alpha = .35, label = 'Кол-во оплат в мес')
ax.set_title('Выручка и платежи')
ax.set_xlabel('Месяцы')
ax.set_ylabel('Выручка, млн.')
ax1.set_ylabel('Платежи, шт.')
ax.legend(loc = 'lower left')
ax1.legend(loc = 'upper right')
ax.grid()
plt.show()
```

```python
##############################################   Задание 2.2   #################################################################

##### Постройте scatter plot для среднего чека для каждого месяца.
##### Для этого создайте новую колонку в сгруппированном датафрейме с отношением суммы платежей к их количеству
##### Сохраните форматирование из предыдущих пунктов
##### Задайте нижнюю границу оси Y как ноль. Верхнюю границу придумайте самостоятельно для лучшего отображения информации.

df_group_month['avg_ch'] = round(df_group_month['sum_pay'] / df_group_month['cnt_pay'], 2)

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(df_group_month['month_pay'], df_group_month['avg_ch'], color = 'gold', edgecolor = 'royalblue', label = 'Ср. чек в мес', s = 100)
ax.set_title('Динамика среднего чека')
ax.set_xlabel('Месяцы')
ax.set_ylabel('Ср. чек')
ax.legend(loc = 'lower left')
ax.grid()
ax.set_ylim([0,6000])
plt.show()
```
